```python
import qiskit as qk
import numpy as np
from qiskit import Aer
from qiskit.visualization import plot_histogram
from qiskit import IBMQ
%matplotlib inline

import matplotlib.pyplot as plt
```

# 2-SAT Problem Instance

Here, we will focus on the problem shown in class:

$f(\vec{x})= (x_1 \lor x_2) \land (\lnot x_2 \lor x_3) \land (\lnot x_1 \lor x_3)$

# Algorithm Construction

```python
num_variables = 3
num_clauses = 3
qr_x = qk.QuantumRegister(num_variables, name='q')
qr_r = qk.QuantumRegister(num_clauses, name='r')
qr_a = qk.QuantumRegister(1, name='a')
cr = qk.ClassicalRegister(num_variables, name='c')
```

## State Preparation

```python
def state_prep():
    circ = qk.QuantumCircuit(qr_x, qr_r, qr_a, cr)
    circ.x(qr_a) # prepares ancilla in |1>
    circ.h(qr_a) # prepares ancilla in |->
    circ.h(qr_x) # prepares data qubits in |+>
    circ.barrier()
    return circ
```

## Oracle

```python
def oracle():
    circ = qk.QuantumCircuit(qr_x, qr_r, qr_a, cr)
    # --- Clause encoding ---
    # Clause 1
    circ.x([qr_x[0], qr_x[1]])
    circ.mcx([qr_x[0], qr_x[1]],qr_r[0])
    circ.x([qr_x[0], qr_x[1]])
    circ.barrier()
    # Clause 2
    circ.x([qr_x[2]])
    circ.mcx([qr_x[1], qr_x[2]],qr_r[1])
    circ.x([qr_x[2]])
    circ.barrier()
    # Clause 3
    circ.x([qr_x[2]])
    circ.mcx([qr_x[0], qr_x[2]],qr_r[2])
    circ.x([qr_x[2]])
    circ.barrier()
    
    # --- Verification of Satisfication ---
    circ.x(qr_r)
    circ.mcx([qr_r[0], qr_r[1], qr_r[2]], qr_a[0])
    circ.x(qr_r)
    circ.barrier()

    # --- Uncomputation ---
    # Clause 3
    circ.x([qr_x[2]])
    circ.mcx([qr_x[0], qr_x[2]],qr_r[2])
    circ.x([qr_x[2]])
    circ.barrier()
    # Clause 2
    circ.x([qr_x[2]])
    circ.mcx([qr_x[1], qr_x[2]],qr_r[1])
    circ.x([qr_x[2]])
    circ.barrier()
    # Clause 1
    circ.x([qr_x[0], qr_x[1]])
    circ.mcx([qr_x[0], qr_x[1]],qr_r[0])
    circ.x([qr_x[0], qr_x[1]])
    circ.barrier()
    
    return circ
```

## Diffuser

```python
def diffuser():
    circ = qk.QuantumCircuit(qr_x, qr_r, qr_a, cr)
    circ.h(qr_x)
    circ.x(qr_x)
    circ.barrier()
    circ.h(qr_x[2])
    circ.mcx([qr_x[0],qr_x[1]],qr_x[2])
    circ.h(qr_x[2])
    circ.barrier()
    circ.x(qr_x)
    circ.h(qr_x)
    circ.barrier()
    return circ
```

# Implementation

## Single Iteration

```python
grovers_2sat = qk.QuantumCircuit(qr_x, qr_r, qr_a, cr)
grovers_2sat.compose(state_prep(), inplace=True)
grovers_2sat.compose(oracle(), inplace=True)
grovers_2sat.compose(diffuser(), inplace=True)
grovers_2sat.measure(qr_x,cr)
grovers_2sat.draw('mpl', fold=-1)
```

```python
backend = Aer.get_backend('qasm_simulator')
job = qk.execute(grovers_2sat, backend=backend, shots=8000)
results = job.result()
counts = results.get_counts(grovers_2sat)
plot_histogram(counts)
```

## Multiple Iterations

```python
def grovers_2sat_iter(iterations):
    circ = qk.QuantumCircuit(qr_x, qr_r, qr_a, cr)
    circ.compose(state_prep(), inplace=True)
    for i in range(iterations):
        circ.compose(oracle(), inplace=True)
        circ.compose(diffuser(), inplace=True)
    circ.measure(qr_x,cr)
    return circ
```

```python
max_iterations = 10
shots = 8000
iteration_counts = []
backend = Aer.get_backend('qasm_simulator')
for i in range(max_iterations):
    print('iteration: ', i)
    grovers_2sat = grovers_2sat_iter(i)
    job = qk.execute(grovers_2sat, backend=backend, shots=shots, optimization_level=0)
    results = job.result()
    counts = results.get_counts(grovers_2sat)
    iteration_counts.append(counts)
```

```python
fig = plt.figure(dpi=90)
ax = fig.add_subplot(111)

bit_values = iteration_counts[0].keys()
for bit_val in bit_values:
    
    bit_prob = []
    for counts in iteration_counts:
        bit_counts = counts[bit_val]
        bit_prob.append(bit_counts/shots)
    
    ax.plot(list(range(max_iterations)), bit_prob, label=bit_val)

ax.legend(loc=1, bbox_to_anchor=(1.3,1))
ax.set_ylabel('Probability', fontsize=15)
ax.set_xlabel('Grover Iteration', fontsize=15)

plt.tight_layout()
```

Focusing on the states $|111\rangle, |110\rangle, |101\rangle$.

```python
fig = plt.figure(dpi=90)
ax = fig.add_subplot(111)

bit_values = ['101', '110', '111']
for bit_val in bit_values:
    
    bit_prob = []
    for counts in iteration_counts:
        bit_counts = counts[bit_val]
        bit_prob.append(bit_counts/shots)
    
    ax.plot(list(range(max_iterations)), bit_prob, label=bit_val)

    
# Optimal Grover Iterations
K = np.ceil(np.pi/4*np.sqrt(2**3))
ax.axvline(x=K, ls='--', color='k', lw=2)
    
ax.legend(loc=1, bbox_to_anchor=(1.3,1))
ax.set_ylabel('Probability', fontsize=15)
ax.set_xlabel('Grover Iteration', fontsize=15)

plt.tight_layout()
```

### Cumulative Probability

```python
fig = plt.figure(dpi=90)
ax = fig.add_subplot(111)

bit_values = ['101', '110', '111']

# Let's plot the sum of probabilities for all expected solutions

sum_bit_prob = np.zeros(max_iterations)
sum_faulty_bit_prob = np.zeros(max_iterations)
for bit_val in bit_values:
    # --- Noiseless Case ---
    # Construct list of probabilities as a function of grover iterations k
    bit_prob = []
    for counts in iteration_counts:
        bit_counts = counts[bit_val]
        bit_prob.append(bit_counts/shots)
    # Sum probabilities
    sum_bit_prob += np.array(bit_prob)


ax.plot(list(range(max_iterations)), sum_bit_prob, label='Noiseless')

ax.legend(loc=1)
ax.set_ylabel('Probability', fontsize=15)
ax.set_xlabel('Grover Iteration', fontsize=15)

plt.tight_layout()
```

# Faulty Grover

```python
from qiskit.providers.aer.noise import NoiseModel
from qiskit.providers.aer.noise import depolarizing_error
```

Now, we will add dephasing noise to all single qubit operations and examine its effect on the 2-SAT problem

```python
# Create an empty noise model
noise_model = NoiseModel()

# Add depolarizing error to the hadamard gates
p = 0.005 # probability of depolarization
error = depolarizing_error(p, 1)
noise_model.add_all_qubit_quantum_error(error, ['h', 'x', 'sx'])
```

We can examine the effect of noise under two assumptions: 

(1) We have the ability to natively implement every multi-qubit gate defined within Grover's Search

(2) We must compile the multi-qubit operations into native one and two-qubit gates.

We will examine both scenarios below.

## Assuming Native Multi-Qubit Gates

```python
max_iterations = 10
shots = 8000
faulty_iteration_counts = []
backend = Aer.get_backend('qasm_simulator')
for i in range(max_iterations):
    print('iteration: ', i)
    grovers_2sat = grovers_2sat_iter(i)
    job = qk.execute(grovers_2sat, backend=backend, shots=shots, noise_model=noise_model,
                    optimization_level=0)
    results = job.result()
    counts = results.get_counts(grovers_2sat)
    faulty_iteration_counts.append(counts)
```

```python
fig = plt.figure(dpi=90)
ax = fig.add_subplot(111)

bit_values = ['101', '110', '111']

# Let's plot the sum of probabilities for all expected solutions

sum_bit_prob = np.zeros(max_iterations)
sum_faulty_bit_prob = np.zeros(max_iterations)
for bit_val in bit_values:
    # --- Noiseless Case ---
    # Construct list of probabilities as a function of grover iterations k
    bit_prob = []
    for counts in iteration_counts:
        bit_counts = counts[bit_val]
        bit_prob.append(bit_counts/shots)
    # Sum probabilities
    sum_bit_prob += np.array(bit_prob)
    
    # --- Noisy Case ---
    faulty_bit_prob = []
    for counts in faulty_iteration_counts:
        bit_counts = counts[bit_val]
        faulty_bit_prob.append(bit_counts/shots)
    # Sum probabilities
    sum_faulty_bit_prob += np.array(faulty_bit_prob)

ax.plot(list(range(max_iterations)), sum_bit_prob, label='Noiseless')
ax.plot(list(range(max_iterations)), sum_faulty_bit_prob, ls='--', label='Faulty')

ax.legend(loc=1, bbox_to_anchor=(1.3,1))
ax.set_ylabel('Probability', fontsize=15)
ax.set_xlabel('Grover Iteration', fontsize=15)

plt.tight_layout()
```

## Assuming Multi-Qubit Gates Need to be Compiled to One and Two-Qubit Gates

We use the Qiskit's built in transpilation feature to compile Grover's Search operations into a particular gate set.

```python
grovers_2sat = grovers_2sat_iter(1)
# Compile into gate set
grovers_2sat_transpiled = qk.transpile(grovers_2sat, backend=backend, 
                                       basis_gates=['sx','x','rz','cx'])
grovers_2sat_transpiled.draw('mpl')
```

Now, let's run a simulation using the transpiled circuit.

```python
max_iterations = 10
shots = 8000
faulty_iteration_counts = []
backend = Aer.get_backend('qasm_simulator')
for i in range(max_iterations):
    print('iteration: ', i)
    grovers_2sat = grovers_2sat_iter(i)
    # Compile into gate set
    grovers_2sat_transpiled = qk.transpile(grovers_2sat, backend=backend, 
                                           basis_gates=['sx','x','rz','cx'])
    job = qk.execute(grovers_2sat, backend=backend, shots=shots, noise_model=noise_model,
                    optimization_level=0)
    results = job.result()
    counts = results.get_counts(grovers_2sat)
    faulty_iteration_counts.append(counts)
```

```python
fig = plt.figure(dpi=90)
ax = fig.add_subplot(111)

bit_values = ['101', '110', '111']

# Let's plot the sum of probabilities for all expected solutions

sum_bit_prob = np.zeros(max_iterations)
sum_faulty_bit_prob = np.zeros(max_iterations)
for bit_val in bit_values:
    # --- Noiseless Case ---
    # Construct list of probabilities as a function of grover iterations k
    bit_prob = []
    for counts in iteration_counts:
        bit_counts = counts[bit_val]
        bit_prob.append(bit_counts/shots)
    # Sum probabilities
    sum_bit_prob += np.array(bit_prob)
    
    # --- Noisy Case ---
    faulty_bit_prob = []
    for counts in faulty_iteration_counts:
        bit_counts = counts[bit_val]
        faulty_bit_prob.append(bit_counts/shots)
    # Sum probabilities
    sum_faulty_bit_prob += np.array(faulty_bit_prob)

ax.plot(list(range(max_iterations)), sum_bit_prob, label='Noiseless')
ax.plot(list(range(max_iterations)), sum_faulty_bit_prob, ls='--', label='Faulty')

ax.legend(loc=1, bbox_to_anchor=(1.3,1))
ax.set_ylabel('Probability', fontsize=15)
ax.set_xlabel('Grover Iteration', fontsize=15)

plt.tight_layout()
```
